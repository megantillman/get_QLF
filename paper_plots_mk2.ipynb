### Load best fit values and organize them.

```python
def organize_set_varied_dicts(set_dict, varied_dict):
    fixed_params = {'Mscrit':set_dict['Transition point'], 'presig':set_dict['Pre-disk sigma'],\
                'postsig':set_dict['Post-disk sigma'], 'preslope': set_dict['Pre-disk slope'],\
                'Mbhnorm':set_dict['Pre-disk normalization'], 'postnorm':10**set_dict['Post-disk normalization']} #allowing no parameters to vary

    varied_params = {'Mscrit':[varied_dict[v]['Transition point'] for v in varied_dict],\
              'presig':[varied_dict[v]['Pre-disk sigma'] for v in varied_dict],\
              'postsig':[varied_dict[v]['Post-disk sigma'] for v in varied_dict],\
              'preslope':[varied_dict[v]['Pre-disk slope'] for v in varied_dict],\
              'Mbhnorm':[varied_dict[v]['Pre-disk normalization'] for v in varied_dict],\
              'postnorm':[10**varied_dict[v]['Post-disk normalization'] for v in varied_dict]}
    
    return fixed_params, varied_params

fixed_fiducial, varied_fiducial = load_obj('fixed_2P-fiducial'), load_obj('varied_2P-fiducial')
fixed_B1, varied_B1 = load_obj('fixed_2P-B1'), load_obj('varied_2P-B1')
fixed_B2, varied_B2 = load_obj('fixed_2P-B2'), load_obj('varied_2P-B2')
fixed_B3, varied_B3 = load_obj('fixed_2P-B3'), load_obj('varied_2P-B3')
fixed_P, varied_P = load_obj('fixed_2P-phantom'), load_obj('varied_2P-phantom')
fixed_L, varied_L = load_obj('fixed_L-fiducial'), load_obj('varied_L-fiducial')
fixed_LB1, varied_LB1 = load_obj('fixed_L-B1'), load_obj('varied_L-B1')
fixed_LB2, varied_LB2 = load_obj('fixed_L-B2'), load_obj('varied_L-B2')
fixed_LB3, varied_LB3 = load_obj('fixed_L-B3'), load_obj('varied_L-B3')

fixed_paramsF, varied_paramsF = organize_set_varied_dicts(fixed_fiducial, varied_fiducial)

fixed_paramsB1, varied_paramsB1 = organize_set_varied_dicts(fixed_B1, varied_B1)

fixed_paramsB2, varied_paramsB2 = organize_set_varied_dicts(fixed_B2, varied_B2)

fixed_paramsB3, varied_paramsB3 = organize_set_varied_dicts(fixed_B3, varied_B3)

fixed_paramsL, varied_paramsL = organize_set_varied_dicts(fixed_L, varied_L)

fixed_paramsLB1, varied_paramsLB1 = organize_set_varied_dicts(fixed_LB1, varied_LB1)

fixed_paramsLB2, varied_paramsLB2 = organize_set_varied_dicts(fixed_LB2, varied_LB2)

fixed_paramsLB3, varied_paramsLB3 = organize_set_varied_dicts(fixed_LB3, varied_LB3)

fixed_paramsP, varied_paramsP = organize_set_varied_dicts(fixed_P, varied_P)
```

# $M_*$-$M_{*bulge}$ Relations (FIGURE B1)

```python
fig = plt.figure(figsize=(7,5))

def BT_alex_fit(Mstar): #Mstar must be in units of solar mass
    BT = 0.4240*(Mstar/1e10)**0.3887 #this is the ratio of bulge stellar mass to total stellar mass
    BT[Mstar < 7e7] = 0
    BT[Mstar > 7e10] = 1
    return BT

def fsigmoid(x, a, b):
    return 1.0 / (1.0 + np.exp(-a*(x-b)))

both = {'x':[], 'y':[], 'yerrup':[], 'yerrdown':[]}

benson = {'x':[], 'y':[], 'yerrup':[], 'yerrdown':[]}
f = open('plot_data/Benson+2017.txt', 'r')
for line in f:
    if line[0] != '#':
        dat = line.split(', ')
        dat[-1] = dat[-1].split('\n')[0]
        dat = [float(d) for d in dat]
        benson['x'].append(float(dat[0]))
        benson['y'].append(float(dat[1]))
        both['x'].append(float(dat[0]))
        both['y'].append(float(dat[1]))
        benson['yerrup'].append(float(dat[2])-float(dat[1]))
        benson['yerrdown'].append(float(dat[1])-float(dat[3]))
        both['yerrup'].append(float(dat[2])-float(dat[1]))
        both['yerrdown'].append(float(dat[1])-float(dat[3]))
        
salo = {'x':[], 'y':[]}
f = open('plot_data/Salo+2015.txt', 'r')
for line in f:
    if line[0] != '#':
        dat = line.split(', ')
        dat[-1] = dat[-1].split('\n')[0]
        dat = [float(d) for d in dat]
        salo['x'].append(float(dat[0]))
        salo['y'].append(float(dat[1]))
        both['x'].append(float(dat[0]))
        both['y'].append(float(dat[1]))
        both['yerrup'].append(0)
        both['yerrdown'].append(0)
        
        
qlf = QLF(1.0, 0.005)
xdata_model = qlf.StellBins
ydata_model = BT_alex_fit(10**xdata_model)

# params, pcov = sp.optimize.curve_fit(fsigmoid, xdata_model, ydata_model, method='dogbox', p0 = [1,9], bounds = ([1,5],[3,10.5]))

xdata_obs = both['x']
ydata_obs = both['y']
params, pcov = sp.optimize.curve_fit(fsigmoid, xdata_obs, ydata_obs, method='dogbox', p0 = [1,9], bounds = ([1,5],[3,10.5]))
plt.plot(xdata_model, fsigmoid(xdata_model, params[0], params[1]), label='B1', c='k')
print('Obs data params:',params)

xdata_obs = both['x']
ydata_obs = np.zeros(len(both['y'])) + np.asarray(both['y']) + np.asarray(both['yerrup'])
params, pcov = sp.optimize.curve_fit(fsigmoid, xdata_obs, ydata_obs, method='dogbox', p0 = [1,9], bounds = ([1,5],[15,10.5]))
plt.plot(xdata_model, fsigmoid(xdata_model, params[0], params[1]), label='B2', c='k', ls = 'dashed')
print('Upper bound params:',params)

xdata_obs = both['x']
ydata_obs = np.zeros(len(both['y'])) + np.asarray(both['y']) - np.asarray(both['yerrdown'])
params, pcov = sp.optimize.curve_fit(fsigmoid, xdata_obs, ydata_obs, method='dogbox', p0 = [1,9], bounds = ([0,5],[3,15]))
plt.plot(xdata_model, fsigmoid(xdata_model, params[0], params[1]), label='B3', c='k', ls = 'dotted')
print('Lower bound params:',params)

plt.errorbar(benson['x'], benson['y'], yerr = [benson['yerrdown'], benson['yerrup']], c='r', fmt='.', label='Benson+2017', capsize=3)
plt.scatter(salo['x'], salo['y'], c='k', marker='x', label='Salo+2015 (median)')

plt.legend()

plt.axvline(10, c='k', linestyle='dotted')

plt.xlabel(r'$\log_{10}[\rmM_*/M_\odot]$')
plt.ylabel('B/T')
plt.xlim([5,12.5])
plt.tight_layout()

plt.savefig('plots/FINAL_paper_plots/Mstar-Mbulge_relation.pdf')
```
